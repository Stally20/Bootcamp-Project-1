```python
import os
import csv
import pandas as pd
nba_combined = os.path.join("nba_2017_players_stats_combined.csv")
nba_pie = os.path.join("nba_2017_pie.csv")
nba_salary = os.path.join("revised_2016-2017_nba_salaries.csv")
#Salary Data Note: If more than 1 salary, took max of multiple teams or same team, or if same used that salary
#Players with Multiple Teams:
#Joffrey Lauvergne - OKC/SA
#Marcelo Huertas - LAL/HOU
```

```python
stats_orig = pd.read_csv(nba_combined)
pie_orig = pd.read_csv(nba_pie)
salary_orig = pd.read_csv(nba_salary)
```

```python
#Using only Players listed on original stats csv as it is a smaller list and will leave out NaN values
stats_pie = pd.merge(stats_orig, pie_orig, how="inner", on="PLAYER")
stats_pie.head()
```

```python
full_nba_orig = pd.merge(stats_pie, salary_orig, how="outer", on="PLAYER")
full_nba = full_nba_orig.set_index("Rk")
full_nba.head()
```

```python
full_nba.columns
```

```python
del full_nba['Unnamed: 0']
del full_nba['MPG']
del full_nba['W_x']
del full_nba['TEAM_y']
del full_nba['AGE_y']
del full_nba['GP_y']
del full_nba['MIN']
del full_nba['EFG%']
del full_nba['PACE_y']
del full_nba['PIE_y']
full_nba.columns
```

```python
full_nba = full_nba.rename(columns={'PLAYER':'Player', 'POSITION_x':'Pos', 'AGE_x':'Age', 'MP':'Min/Gm', 'FG':'FGM/Gm', 
       'FGA':'FGA/Gm', '3P':'3PtM/Gm', '3PA':'3PtA/Gm', '2P':'2PtM/Gm', '2PA':'2PtA/Gm', 'FT':'FTM/Gm', 'FTA':'FTA/Gm',
       'ORB':'O-Reb/Gm', 'DRB':'D-Reb/Gm', 'TRB':'Reb/Gm', 'AST':'Ast/Gm', 'STL':'Stl/Gm', 'BLK':'Blk/Gm', 'TOV':'TO/Gm',
       'PF':'Fouls/Gm', 'POINTS':'Pts/Gm','TEAM_x':'Team', 'GP_x':'Games', 'ORPM':'Off +/-', 'DRPM':'Def +/-',
       'RPM':'Real +/-', 'WINS_RPM':'Wins +/-', 'PIE_x':'PIE', 'PACE_x':'Pace', 'W_y':'W', 'OFFRTG':'Off Rtg',
       'DEFRTG':'Def Rtg', 'NETRTG':'Net Rtg', 'AST%':'Ast%', 'AST/TO':'Ast/TO', 'ASTÂ RATIO':'Ast Ratio', 'OREB%':'O-Reb%',
       'DREB%':'D-Reb%', 'REB%':'Reb%', 'TOÂ RATIO':'TO Ratio', 'TS%':'True Shooting%', 'USG%':'Usage%', 'SALARY':'Salary'})
full_nba.head()
```
